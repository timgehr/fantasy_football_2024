```python
import nfl_data_py as nfl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
weekly_table = nfl.import_weekly_data([2023, 2022, 2021])
yearly_table = nfl.import_seasonal_data([2023, 2022, 2021])
weekly = pd.DataFrame(weekly_table)
yearly = pd.DataFrame(yearly_table)

weekly['half_ppr'] = (weekly['fantasy_points'] + weekly['fantasy_points_ppr']) / 2
yearly['half_ppr'] = (yearly['fantasy_points'] + yearly['fantasy_points_ppr']) / 2

wr_weekly = weekly[weekly['position_group'] == 'WR']
wr_list = wr_weekly[["player_id", "player_name", "season", "recent_team"]].drop_duplicates()
wr_stats = pd.merge(yearly, wr_list, how="inner", on=['player_id', 'season'])

# 2022 Projections by Sleeper
wr_proj_2022 = pd.read_csv('2022_Sleeper_Projections.csv')
wr_proj_2022['projected_points_per_game'] = wr_proj_2022['projected_points'] / 17
wr_proj_2022['projected_points_per_reception'] = wr_proj_2022['projected_points'] / wr_proj_2022['projected_rec']
wr_proj_2022['projected_rec_tds_per_game'] = wr_proj_2022['projected_rec_tds'] / 17
wr_proj_2022['projected_points_per_touchdown'] = wr_proj_2022['projected_points'] / (wr_proj_2022['projected_rec_tds'] + 1)
wr_proj_2022['projected_rec_per_game'] = wr_proj_2022['projected_rec'] / 17
wr_proj_2022['projected_rec_yds_per_game'] = wr_proj_2022['projected_rec_yds'] / 17
wr_proj_2022 = wr_proj_2022.add_suffix('_2022')
wr_proj_2022['player_name'] = wr_proj_2022['player_name_2022']
# 2023 Projections by Sleeper
wr_proj_2023 = pd.read_csv('SleeperPredictions.csv')
wr_proj_2023['projected_points_per_game'] = wr_proj_2023['projected_points'] / 17
wr_proj_2023['projected_points_per_reception'] = wr_proj_2023['projected_points'] / wr_proj_2023['projected_rec']
wr_proj_2023['projected_rec_tds_per_game'] = wr_proj_2023['projected_rec_tds'] / 17
wr_proj_2023['projected_points_per_touchdown'] = wr_proj_2023['projected_points'] / (wr_proj_2023['projected_rec_tds'] + 1)
wr_proj_2023['projected_rec_per_game'] = wr_proj_2023['projected_rec'] / 17
wr_proj_2023['projected_rec_yds_per_game'] = wr_proj_2023['projected_rec_yds'] / 17
wr_proj_2023 = wr_proj_2023.add_suffix('_2023')
wr_proj_2023['player_name'] = wr_proj_2023['player_name_2023']
# 2024 Projections by FantasyPros
wr_proj_2024 = pd.read_csv('FantasyPros_Fantasy_Football_Projections_WR.csv')
wr_proj_2024 = wr_proj_2024.add_suffix('_2024')
wr_proj_2024['player_name'] = wr_proj_2024['Player_2024'].str.get(0) + '.' + wr_proj_2024['Player_2024'].str.split(' ').str[1]
wr_proj_2024['projected_points_per_game_2024'] = wr_proj_2024['FPTS_2024'] / 17
wr_proj_2024['projected_points_per_reception_2024'] = wr_proj_2024['FPTS_2024'] / wr_proj_2024['REC_2024']
wr_proj_2024['projected_rec_tds_per_game_2024'] = wr_proj_2024['TDS_2024'] / 17
wr_proj_2024['projected_points_per_touchdown_2024'] = wr_proj_2024['FPTS_2024'] / (wr_proj_2024['TDS_2024'] + 1)
wr_proj_2024['projected_rec_per_game_2024'] = wr_proj_2024['REC_2024'] / 17
wr_proj_2024['projected_rec_yds_per_game_2024'] = wr_proj_2024['YDS_2024'] / 17
wr_proj_2024['projected_ppg_rank_2024'] = wr_proj_2024['FPTS_2024'].rank(ascending=False)

# Merge projections with stats
wr_stats = pd.merge(wr_proj_2022, wr_stats, how="inner", on=['player_name'])
wr_stats = pd.merge(wr_proj_2023, wr_stats, how="inner", on=['player_name'])
wr_stats = pd.merge(wr_proj_2024, wr_stats, how="inner", on=['player_name'])
wr_stats['TIBScore'] = 0


wr_stats['ppg'] = wr_stats['half_ppr'] / wr_stats['games']

wr_stats_2021 = wr_stats[wr_stats['season'] == 2021]
wr_stats_2021['ppg_rank'] = wr_stats_2021['ppg'].rank(ascending=False)
wr_stats_2022 = wr_stats[wr_stats['season'] == 2022]
wr_stats_2022['ppg_rank'] = wr_stats_2022['ppg'].rank(ascending=False)
wr_stats_2023 = wr_stats[wr_stats['season'] == 2023]

stat_labels = ['receptions', 'targets', 'receiving_yards', 'receiving_tds', 'receiving_air_yards',
       'receiving_yards_after_catch', 'receiving_first_downs', 'receiving_epa',
       'receiving_2pt_conversions', 'racr', 'target_share', 'air_yards_share',
       'wopr_x','tgt_sh', 'ay_sh', 'yac_sh', 'wopr_y', 'ry_sh',
       'rfd_sh', 'rtdfd_sh']
```

```python
for label in stat_labels:
    print(label, np.corrcoef(wr_stats_2023[label], wr_stats_2023['ppg'])[0, 1])
```

```python
wr_stats_add_2022 = wr_stats_2022.add_suffix('_2022')
wr_stats_add_2022['player_id'] = wr_stats_add_2022['player_id_2022']
wr_stats_test_22_23 = pd.merge(wr_stats_2023, wr_stats_add_2022, how="inner", on=['player_id'])

wr_stats_add_2021 = wr_stats_2021.add_suffix('_2021')
wr_stats_add_2021['player_id'] = wr_stats_add_2021['player_id_2021']
wr_stats_test_21_22 = pd.merge(wr_stats_2022, wr_stats_add_2021, how="inner", on=['player_id'])

best_correlation = 0
best_correlation_index = [0,0,0,0]
best_correlation_stat_e = ''
max_coef = 7
correlation_score = np.zeros((max_coef, max_coef, max_coef, max_coef, max_coef))
for label in stat_labels:
    for a in range(0, max_coef):
        for b in range(0, max_coef):
            for c in range(0, max_coef):
                for d in range(0, max_coef):
                    wr_stats_test_22_23['TIBScore'] = (wr_stats_test_22_23[label + '_2022'] ** (a/4) + 
                                                wr_stats_test_22_23['projected_rec_per_game_2023'] ** (b/4) + 
                                                (wr_stats_test_22_23['projected_rec_tds_per_game_2023'] * 6) ** (c/4) +
                                                wr_stats_test_22_23['tgt_sh_2022'] ** (d/4))
                    wr_stats_test_21_22['TIBScore'] = (wr_stats_test_21_22[label + '_2021'] ** (a/4) + 
                                                wr_stats_test_21_22['projected_rec_per_game_2022'] ** (b/4) + 
                                                (wr_stats_test_21_22['projected_rec_tds_per_game_2022'] * 6) ** (c/4) +
                                                wr_stats_test_21_22['tgt_sh_2021'] ** (d/4))
                    correlation_score[a,b,c,d] = (np.corrcoef(wr_stats_test_22_23['TIBScore'], wr_stats_test_22_23['ppg'])[0, 1] + np.corrcoef(wr_stats_test_21_22['TIBScore'], wr_stats_test_21_22['ppg'])[0, 1]) / 2
    print(label + ' done')
    correlation_score[np.isnan(correlation_score)] = 0
    max_value = np.max(correlation_score)
    max_index = np.unravel_index(np.argmax(correlation_score), correlation_score.shape)
    if max_value > best_correlation:
        best_correlation = max_value
        for i in range(0, 4):
            best_correlation_index[i] = max_index[i].item()
        best_correlation_stat_e = label

print(best_correlation)
print(str(best_correlation_index))
print('with stat e as ' + best_correlation_stat_e)
```
